### increment salary of emplyee by 5 % if saray 3500

```python
salary=np.random.randint(25000,55000,10)
```

```python
salary
```

```python
np.where(salary>35000,salary*0.05+salary,salary).astype('int')
```

```python
np.where(salary<35000,salary*1.03,np.where((salary>35000) & (salary<45000),salary*1.08,salary*1.13)).astype('int')
```
