```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
driver=webdriver.Chrome()
driver.get(r'https://www.imdb.com/chart/top/')
```

```python
obj=driver.find_element(By.XPATH,f'/html/body/div[2]/main/div/div[3]/section/div/div[2]/div/ul/li[248]/div/div/div/div/div[1]/div/div[1]/img').get_attribute('src')
print(obj)
```

```python
mname=[]
year=[]
duration=[]
rating=[]
rpeople=[]
crating=[]

for i in range(1,251):
    title=driver.find_element(By.XPATH,f'/html/body/div[2]/main/div/div[3]/section/div/div[2]/div/ul/li[{i}]/div/div/div/div/div[2]/div[2]/a/h3').text
    mname.append(title)

    dtime=driver.find_element(By.XPATH,f'/html/body/div[2]/main/div/div[3]/section/div/div[2]/div/ul/li[{i}]/div/div/div/div/div[2]/div[3]/span[1]').text
    year.append(dtime)
    
    dutime=driver.find_element(By.XPATH,f'/html/body/div[2]/main/div/div[3]/section/div/div[2]/div/ul/li[{i}]/div/div/div/div/div[2]/div[3]/span[2]').text
    duration.append(dutime)

    rtime=driver.find_element(By.XPATH,f'/html/body/div[2]/main/div/div[3]/section/div/div[2]/div/ul/li[{i}]/div/div/div/div/div[2]/span/div/span/span[1]').text
    rating.append(rtime)

    rpep=driver.find_element(By.XPATH,f'/html/body/div[2]/main/div/div[3]/section/div/div[2]/div/ul/li[{i}]/div/div/div/div/div[2]/span/div/span/span[2]').text
    rpeople.append(rpep)
    try:
        cret=driver.find_element(By.XPATH,f'/html/body/div[2]/main/div/div[3]/section/div/div[2]/div/ul/li[{i}]/div/div/div/div/div[2]/div[3]/span[3]').text
        
    except:
        cret='blank'
    crating.append(cret)
```

```python
dicts={'movie_name':mname,'released year':year,'duration':duration,'rating':rating,'people rated':rpeople,'Rating':crating}
dicts['Rating']
```

```python
df=pd.DataFrame(dicts)
df
```

```python
imgs=[]
movie_l=[]
for i in range(1,251):
    img_path=driver.find_element(By.XPATH,f'/html/body/div[2]/main/div/div[3]/section/div/div[2]/div/ul/li[{i}]/div/div/div/div/div[1]/div/div[1]/img').get_attribute('src')
    imgs.append(img_path)
    img_l=driver.find_element(By.XPATH,f'/html/body/div[2]/main/div/div[3]/section/div/div[2]/div/ul/li[{i}]/div/div/div/div/div[2]/div[2]/a').get_attribute('href')
    movie_l.append(img_l)
```

```python
df['Img_link']=imgs
df['Movie_Link']=movie_l
```

```python
df.to_csv('E:\Datascience\dataset\IMDb.csv')
```
