## marker, markersize, 

```python
plt.figure(figsize=(13,5))

plt.title("virat kohli and rohit sharma Carrer Comaparisions ")
plt.xlabel("season")
plt.ylabel("runs")

plt.xticks(dfs['index'])
plt.yticks([i for i in range(100,1100,100)])

plt.plot(dfs['index'],dfs['RG Sharma'],marker='o',markersize=6,mec='red',markerfacecolor='green',color='red')
plt.plot(dfs['index'],dfs['V Kohli'],marker='s',markersize=6,mec='red',markerfacecolor='green',color='green')
```

# plt.grid() 

```python
plt.figure(figsize=(13,5))

plt.title("virat kohli and rohit sharma Carrer Comaparisions ")
plt.xlabel("season")
plt.ylabel("runs")

plt.xticks(dfs['index'])
plt.yticks([i for i in range(100,1100,100)])


# plt.grid()

plt.plot(dfs['index'],dfs['RG Sharma'],marker='o',
         markersize=6,color='red',label="rohit",linestyle='dashed')

plt.plot(dfs['index'],dfs['V Kohli'],marker='s',
         markersize=6,color='green',label="virat",linestyle=':')

plt.legend(loc="upper left")
```
