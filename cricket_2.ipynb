## import libray

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
# /html/body/section[2]/div/div/div[2]/div/div[1]

# /html/body/section[2]/div/div/div[2]/div/div[2]
```

```python
driver=webdriver.Chrome()
players={}
for link in links[:5]:
    driver.get(link)
    time.sleep(5)
    driver.execute_script('window.scrollTo(0,document.body.scrollHeight);')
    time.sleep(3)
    player_info=driver.find_element(By.XPATH,'/html/body/section[2]/div/div/div[2]/div/div[1]').text
    # print(player_info)
    pro_info=driver.find_element(By.XPATH,'/html/body/section[2]/div/div/div[2]/div/div[2]').text
    # print(pro_info)
    p_name=driver.title
    # print(p_name)
    players[p_name]={'player_info':player_info,'pro_info':pro_info}
    # print()
    # print('_'*100)
    # print()
```
