```python
import pandas as pd
import pickle

with open(r'E:\Datascience\DATA\share_markets.pkl','rb')as file:
    data=pickle.load(file)
```

```python
data=data.split('\n')
```

```python
df = pd.DataFrame(data, columns=["Raw_Data"])

print(df)
```

```python
# Remove the first row (header inside data)
df = df.iloc[1:].reset_index(drop=True)

print(df)
```

```python
# Extract Serial Number (first number before space)
df["S.N."] = df["Raw_Data"].str.extract(r'^(\d+)')

```

```python
# Extract DP ID (second number)
df["DP_ID"] = df["Raw_Data"].str.extract(r'^\d+\s+(\d+)')
```

```python
df.head()
```

```python
# Remove first two numbers from Raw_Data
df["Clean_Data"] = df["Raw_Data"].str.replace(r'^\d+\s+\d+\s+', '', regex=True)

df[["Clean_Data"]].head()
df
```

```python
# Extract telephone numbers
df["Telephone"] = df["Clean_Data"].str.extract(r'((?:\d{2}-\d{6,7}(?:,\s*)?)*)$')

df[["Clean_Data", "Telephone"]].head()
```

```python
# Remove telephone from Clean_Data
df["Clean_Data"] = df["Clean_Data"].str.replace(r'((?:\d{2}-\d{6,7}(?:,\s*)?)*)$', '', regex=True).str.strip()

df[["Clean_Data", "Telephone"]].head()
```

```python
# Split address from right side at last comma
df[["Address", "City"]] = df["Clean_Data"].str.rsplit(",", n=1, expand=True)

df["Address"] = df["Address"].str.strip()
df["City"] = df["City"].str.strip()

df[["Address", "City"]].head()
```

```python
# Extract Company Name (everything ending with Limited)
df["Company_Name"] = df["Address"].str.extract(r'^(.*?Limited)')

# Extract actual street address (text after Limited)
df["Street_Address"] = df["Address"].str.replace(r'^.*?Limited', '', regex=True).str.strip()

df[["Company_Name", "Street_Address"]].head()
```

```python
final_df = df[["S.N.", "DP_ID", "Company_Name", "Street_Address", "City", "Telephone"]]

final_df.head()
```

```python
# Combine Street_Address and City into one column called Address
final_df["Address"] = final_df["Street_Address"] + ", " + final_df["City"]

# Drop old columns if you don't need them
final_df = final_df.drop(columns=["Street_Address", "City"])

final_df.head()
```
