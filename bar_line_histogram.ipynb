```python
def bar_plot(year):
    year_data=df[df['year']==year]
    year_data['pop_in_m']=(year_data['pop']/1000000).apply(lambda x : round(x,2))
    year_data=pd.DataFrame(year_data.groupby('continent')['pop_in_m'].sum()).reset_index()
    plt.figure(figsize=(12,5))
    plt.bar(year_data['continent'],year_data['pop_in_m'],width=0.5,color='green')
    plt.show()
        
```

```python
bar_plot(1952)
```
