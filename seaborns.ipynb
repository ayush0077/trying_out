```python
total = df['Value'].sum()
plt.figure(figsize=(10,5))
ax=sns.histplot(df,x='total_bill',bins=[0,5,10,15,20,25,30,35,40,45,50,55,60],hue="sex",stat="percent")

plt.xticks([0,5,10,15,20,25,30,35,40,45,50,55,60])

for container in ax.containers:
    # Get the data values from the container
    bar_values = container.datavalues
    
    # Generate labels with rounded percentages
    labels = [f'{((v / total) * 100):.3f}%' for v in bar_values] # .1f for one decimal place rounding
    
    # Use Matplotlib's bar_label to place the labels
    ax.bar_label(container, labels=labels, label_type='edge')

plt.show()
```
