```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver=webdriver.Chrome()
driver.get('https://cricketnepal.org.np/squads/nepal-mens-senior-national-team/')
```

```python
player_links=[]

for i in range(1,34):
    data=driver.find_element(By.XPATH,f'/html/body/section[3]/div/div/div[2]/div/div[1]/div/a[{i}]').get_attribute('href')
    player_links.append(data)
```

## Pickle Fule 

```python
import pickle # array, df, list, dict, changes to binary form
```

## dump the file in binary form

```python
with open(r'E:\Datascience\Player_links\p_link.pkl','wb')as file:
    pickle.dump(player_links,file) # dump write a binary opbject
    
```
